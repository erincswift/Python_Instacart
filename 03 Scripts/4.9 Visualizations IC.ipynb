## 4.3 Hour of Day Histogram

```python
#Create histogram
hist_how = df_cust_orders_products['order_hour_of_day'].plot.hist(bins = 24)
```

```python
#Export figure
hist_how.figure.savefig(os.path.join(project_path, '04 Analysis', '04 03 Visualizations', 'hist_order_hour.png'))
```

Based on the visualization of customer spending throughout the day, we can conclude that customers order the most between 10:00 AM and 4:00 PM, with 10:00-11:00 AM as the hour with the most orders. Conversely, the early morning hours are the least used hours for sales.

## 4.7 Age and Income Scatterplot

What we learned in this scatter plot is that there are a significant number of customers whose income is 200,000 or less.  We also can see an increase in income at age 40.  The maximum income increases at age 40, as does the number of customers who have an income between 200,000 and 300,000.
