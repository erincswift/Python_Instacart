## 2.0 Exploring Relationships

### 2.1 Create Subset

```python
# Create subset of only numberical fields
sub = df[['order_day_of_week', 'order_hour_of_day', 'days_since_prior_order', 'product_id', 
          'add_to_cart_order', 'reordered', 'aisle_id', 'department_id', 'prices', 'max_orders',
         'mean_price', 'median_days_since', 'age', 'num_dependants', 'income']]
```

```python
sub.shape
```

### 2.2 Correlations

#### Correlation table using pandas

```python
# Create a correlation matrix using subset

sub.corr()
```

#### Correlation heatmap using matplot

```python
# Matplot Correlation Heatmap

matcorf = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=matcorf.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=90) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Save image
matcorf.figure.savefig(os.path.join(path, '04 Analysis', '04 03 Visualizations', 'mat_cor_heatmap.png'))
```

#### Correlation heatmap using seaborn

```python
# Create a subplot with matplotlib
seacorf,ax = plt.subplots(figsize=(15,15))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above
corr = sns.heatmap(sub.corr(), annot = True, ax = ax, fmt = '.3f')
```

#### Top correlations (desc):
        median_days_since/days_since_prior_order = 0.632
        median_days_since/max_orders = -0.628
        max_orders/days_since_prior_order = -0.487
        income/age = 0.417

```python
# Save seaborn correlation matrix
seacorf.figure.savefig(os.path.join(path, '04 Analysis', '04 03 Visualizations', 'sea_cor_heatmap.png'))
```

#### 2.2 Pair Plots

```python
# Create subset of greatest correlation variables
sub_2 = df[['median_days_since', 'days_since_prior_order', 'max_orders','mean_price', 'income', 'age', 'reordered']]
```

```python
#Create random assignments for each record
np.random.seed(4)
dev = np.random.rand(len(sub_2)) <= 0.9
```

```python
#Separate samples into buckets
big = sub_2[dev]
small = sub_2[~dev]
```

```python
#Reducing data frame to only information that is needed
sub_sample = small
```

```python
# Create pair plot 
seapairplot = sns.pairplot(sub_sample)
```

```python
# Save seaborn correlation matrix
seapairplot.figure.savefig(os.path.join(path, '04 Analysis', '04 03 Visualizations', 'sea_pairplot.png'))
```
